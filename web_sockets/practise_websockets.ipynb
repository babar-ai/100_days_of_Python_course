## How WebSockets Work Step-by-Step

#### Step 1: TCP Connection

Before anything else, a TCP connection is established.

- Client ---------------- Server
           - TCP

TCP performs the 3-way handshake:

Client          Server

SYN ----------->
      <-------- SYN-ACK

ACK ----------->

Now a reliable connection exists.


#### Step 2: HTTP Upgrade Request

The client initially sends a normal HTTP request:

GET /ws HTTP/1.1

Upgrade: websocket
Connection: Upgrade

Meaning:

"I want to switch from HTTP to WebSocket."

#### Step 3: Server Accepts Upgrade

Server responds:

HTTP/1.1 101 Switching Protocols

Upgrade: websocket
Connection: Upgrade

Meaning:

"Okay, let's use WebSocket."

At this point:

HTTP ends
WebSocket begins

The same TCP connection remains open.

#### Step 4: Persistent Connection

Instead of:

Request
Response
Connection Closed

we now have:

Client <===============> Server

A single long-lived connection.


#### Step 5: Two-Way Communication

Both sides can send data anytime.

Client ---> Server

Server ---> Client

Client ---> Server

Server ---> Client

This is called:

Full Duplex Communication

Unlike HTTP, the server doesn't need to wait for a request before sending data.

## FastAPI WebSocket Server

```python
from fastapi import FastAPI, WebSocket

app = FastAPI()

@app.websocket("/ws")
async def websocket_endpoint(websocket: WebSocket):

    await websocket.accept()

    while True:
        data = await websocket.receive_text()

        print("Recevied", data)

        await websocket.send_text(
            f"Server received: {data}"
        )
```
