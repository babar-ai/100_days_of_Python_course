## What is an Enum?

- Enum = Enumeration → a set of fixed, constant choices.
= Think of it as a list of labels you can use instead of plain strings or numbers.

It makes your code:

- Clearer (you know all possible options).

- Safer (you can’t accidentally use a wrong value).

```python
# example without Enum

# Using strings
light = "green"                        # Problem: What if someone writes "gren" (typo)? The code breaks silently.

if light == "green":
    print("Go")
elif light == "yellow":
    print("Slow down")
elif light == "red":
    print("Stop")
```

```python
from enum import Enum

class Traffic_Signal(Enum):

    RED = "red"
    GREEN = "green"
    YELLLOW = "yellow"

# now lets use it 

light = Traffic_Signal.RED

if light== Traffic_Signal.RED:
    print("stop!")

elif light== Traffic_Signal.GREEN:
    print("GO!")

else:
    print("ride slowly")
```
